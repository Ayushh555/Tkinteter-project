```python
import tkinter as tk

root = tk.Tk()
root.title("Click Game")
root.geometry("250x200")

count = 0
time_left = 10

def click():
    global count
    if time_left > 0:
        count += 1
        label.config(text=f"Score: {count}")

def countdown():
    global time_left
    if time_left > 0:
        time_left -= 1
        timer_label.config(text=f"Time: {time_left}")
        root.after(1000, countdown)
    else:
        if count >= 20:
            result_label.config(text="You Win!", fg="green")
        else:
            result_label.config(text="You Lose!", fg="red")
        button.config(state="disabled")

label = tk.Label(root, text="Score: 0", font=("Arial", 14))
label.pack(pady=5)

timer_label = tk.Label(root, text="Time: 10", font=("Arial", 12))
timer_label.pack()

button = tk.Button(root, text="Click Me!", command=click)
button.pack(pady=10)

result_label = tk.Label(root, text="", font=("Arial", 14))
result_label.pack()

countdown()

root.mainloop()
```
